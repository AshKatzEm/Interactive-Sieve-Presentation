## More specifically, many attempted, and some still do attempt to capture the infinity of twin primes as a sifted set using increasingly complex exclusions and inclusions, but no one has been directly successful so far.

### Brun’s theorem: the sum of the reciprocals of the twin primes converges. The set of twin primes is relatively tiny.

<br>
<br>
<br>
<br>
<br>

### Some near-miss proofs:

<br>

#### *Chen’s theorem: there are infinitely many primes p such that
#### p + 2 
#### is either a prime 
#### Or 
#### a semiprime (the product of two primes);
#### **This is almost the twin prime conjecture.**

<br>

#### *Chen-Jingrun theorem: every sufficiently large even number is the sum of a prime and another number which is either a prime or a semiprime;
#### **Almost Goldbach’s strong conjecture.**

<br>
<br>
<br>
<br>

### The Friedlander–Iwaniec theorem: there are infinitely many primes of the form  $a^2 + b^4 = p$

```python
# This is factoring using Fermat's method
n = int(input('enter the number you want to factor\n'))
start_time = time.time()
ns = math.ceil(math.sqrt(n)) # we will start from the number above the square of number we want to factor
print('the square root of', n, ' rounded up is:', ns)

for i in range(1, n):
    bs = (ns+i)**2 - n # this is the b^2 term
    print((ns + i), '^2 - ', n, ' = ', bs, sep='')
    
    if math.sqrt(bs).is_integer() is True: # if we have found a b which is a perfect square
        print('(',(ns +i), '+', math.sqrt(bs), ')','(',(ns +i), '-', math.sqrt(bs), ') = (',(ns +i) + math.sqrt(bs), ' , ', (ns +i) - math.sqrt(bs), ')')
        break
    else:
        print(bs, ' is not a perfect square \n')
    if bs > 100000000:
        break

print("--- %s seconds ---" % (time.time() - start_time))
```

```python

def primefactors(n):
    # even number divisible
    npf = list()

    while n % 2 == 0:
        npf.append(2),
        n = n / 2

    # n became odd
    for i in range(3, int(math.sqrt(n)) + 1, 2):

        while (n % i == 0):
            npf.append(i)
            n = n / i

    if n > 2:
        npf.append(n)

    return npf

#999999937
#539873

bound = int(input('enter a b-smooth bound\n'))
n = int(input('enter the number you want to factor\n'))
start_time = time.time()
ns = math.ceil(math.sqrt(n))
print('the square root of', n, ' rounded up is:', ns)
for i in range(1, n):
    bs = (ns+i)**2 - n

    if max(primefactors(bs)) <= bound:
        print((ns + i), '^2 - ', n, ' = ', bs, sep='')
        print('the prime factors of', bs,'are:', end='')

        countsnpf = {}
        for item in primefactors(bs):
            if item in countsnpf:
                countsnpf[item] += 1
            else:
                countsnpf[item] = 1
        print(*[str(k) + '^' + str(v) + ',' for k,v in countsnpf.items()], '\n')
        print('\n')
    if len(str(bs)) > (len(str(n))+2):
        break
print("--- %s seconds ---" % (time.time() - start_time))
```
